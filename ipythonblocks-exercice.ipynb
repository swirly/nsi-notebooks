```python
from ipythonblocks import BlockGrid
import random
```

```python
grille = BlockGrid(10,10,fill=(0,0,0))
for n in range(500):
    couleur=(random.randrange(256),random.randrange(256),random.randrange(256))
    bloc=grille[random.randrange(10),random.randrange(10)]
    bloc.rgb=couleur
    grille.flash(0.01)
```

```python
grille = BlockGrid(10,10,fill=(0,0,0))

def placement():
    x=random.randrange(10)
    y=random.randrange(10)
    direction = random.randrange(1,8)
    couleur =(random.randrange(256),random.randrange(256),random.randrange(256))
    return x,y,direction,couleur

x,y,direction,couleur=placement()
compteur=0

while compteur <50 :
    for b in grille :
        b.rgb=(0,0,0)
    if  direction == 1 :
        x+=1
    elif direction == 2 :
        x+=1
        y+=1
    elif direction == 3 :
        y+=1
    elif direction == 4 :
        y+=1
        x-=1
    elif direction == 5 :
        x-=1
    elif direction == 6 :
        x-=1
        y-=1
    elif direction == 7 :
        y-=1
    elif direction == 8 :
        y-=1
        x+=1
    if x<0 or y<0 or x>9 or y>9 :
        x,y,direction,couleur=placement()
    bloc=grille[x,y]
    bloc.rgb=couleur
    grille.flash(0.1)
    compteur +=1
```

```python
grille = BlockGrid(20,7,fill=(0,0,0))

def placement():
    x=random.randrange(grille.width)
    y=random.randrange(grille.height)
    direction = random.randrange(1,8)
    couleur =(random.randrange(256),random.randrange(256),random.randrange(256))
    return x,y,direction,couleur

x,y,direction,couleur=placement()
compteur=0

while compteur <50 :
    x0=x
    y0=y
    grille[y0,x0].rgb=(0,0,0)
    if  direction == 1 :
        x+=1
    elif direction == 2 :
        x+=1
        y+=1
    elif direction == 3 :
        y+=1
    elif direction == 4 :
        y+=1
        x-=1
    elif direction == 5 :
        x-=1
    elif direction == 6 :
        x-=1
        y-=1
    elif direction == 7 :
        y-=1
    elif direction == 8 :
        y-=1
        x+=1
    if x<0 or y<0 or x>grille.width-1 or y>grille.height-1 :
        x,y,direction,couleur=placement()
    bloc=grille[y,x]
    bloc.rgb=couleur
    grille.flash(0.1)
    compteur +=1
```
