# Les boucles while

Nous avons vu l'instruction `for` qui permet de répéter une même opération sur tous les éléments d'une liste. Il existe un autre type de répétition en Python : la boucle `while`.

While en anglais signifie "Tant que". Cette boucle est liée à une condition. Le fonctionnement est le suivant : tant que la condition est vraie, on répète le bloc d'instruction de la boucle.

**ATTENTION !**

*Si on veut que le programme continue, il faut donc que le bloc d'instruction de la boucle modifie la condition pour qu'elle soit fausse à un moment, sinon on risque une boucle sans fin.*

Voici un exemple simple que vous pouvez exécuter

```python
a=1
while (a<256):
    print(str(a)+" est une puissance de 2 qui est encore contenue dans un octet")
    a = a *2
```
